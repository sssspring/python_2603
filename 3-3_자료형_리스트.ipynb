## 4. 리스트 관련 함수

```python
a =[1,2,3]   # append(x)는 리스트의 맨 마지막에 x를 추가하는 함수이다.
```

```python
a.append(4)
```

```python
a
```

```python
a.append([5,6])  # 리스트 안에는 어떤 자료형도 추가할 수 있다.
```

### sort() 리스트 정렬

```python
a = [1,4,3,2]  # sort 함수는 리스트의 요소를 순소대로 정렬해 준다.
```

```python
a.sort()
```

```python
a
```

```python
a = ['a','c','b']  #문자 역시 알파벳 순서로 정렬 할 수 있다.
```

```python
a.sort()
```

```python
a
```

### reversel() 리스트 뒤집기

```python
a = ['a','c','b']   #reverse 함수는 리스트를 역순으로 뒤집어 준다. ** 리스트 요소들을 순소대로 정렬한 다음 다시 역순으로 정렬하는 것이 아니라 현재의 리스트를 그대로 거꾸로 뒤집는다.
```

```python
a.reverse()
```

```python
a
```

### index(value) 인덱스 반환

```python
a = ['a','c','b']  #index(x) 함수는 리스트에 x 값이 있으면 x의 인덱스 값(위칫값)을 리턴한다.
```

```python
a.index('b')
```

```python
a.index('a')
```

```python
a.index('d')   # 값 0은 a 리스트에 존재하지 않기 때문에 오류가 발생한다.
```

### insert(index,value) 리스트에 요소 삽입

```python
a = [1,2,3   #insert(a, b)는 리스트의 a번째 위치에 b를 삽입하는 함수이다. 
```

```python
a.insert(0,4)
```

```python
a
```

```python
a.insert(3,5)   #리스트 a의 a[3], 즉 네 번째 요소 위치에 값 5를 삽입하라는 뜻이다.
```

```python
a
```

### remove(value) 리스트 요소 제거

```python
a = [1,2,3,1,2,3]    #remove(x)는 리스트에서 첫 번째로 나오는 x를 삭제하는 함수이다.
```

```python
a.remove(3)
```

```python
a
```

```python
a.remove(3)  #remove(3)을 한 번 더 실행하면 다시 3이 삭제된다.
```

```python
a
```

### pop(index) 리스트 요소 끄집어 내기

```python
a = [1,2,3]  # 원본데이터 변경됨!  pop()은 리스트의 맨 마지막 요소를 리턴하고 그 요소는 삭제한다.
```

```python
a.pop()
```

```python
a
```

```python
a = [1,2,3]   #pop(x)는 리스트의 x번째 요소를 리턴하고 그 요소는 삭제한다.
```

```python
a.pop(1)
```

```python
a
```

### count(value) 리스트에 포함된 요소 x의 개수 세기

```python
a = [1,2,3,1]    # count(x)는 리스트 안에 x가 몇 개 있는지 조사하여 그 개수를 리턴하는 함수이다.
```

```python
a.count(1)
```

```python
a.extend([4,5])
```

```python
b = [6,7]
```
