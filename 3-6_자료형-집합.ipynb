## 3. 교집합, 합집합, 차집합 구하기

```python
# set 자료형을 정말 유용하게 사용하는 경우는 교집합, 합집합, 차집합을 구할 때이다.
#먼저 다음과 같이 2개의 set 자료형을 만든 후 따라 해 보자.
```

```python
s1 = set([1,2,3,4,5,6])
s2 = set ([4,5,6,7,8,9])
```

### 교집합 구하기

```python
s1 & s2         #s1과 s2의 교집합을 구해 보자. 다음과 같이 ‘&’를 이용하면 교집합을 간단히 구할 수 있다.
```

```python
s1.intersection(s2)  #intersection 함수를 사용
```

### 합집합 구하기

```python
s1 | s2        #‘|’를 사용하면 합집합을 구할 수 있다. 이때 4, 5, 6처럼 중복해서 포함된 값은 1개씩만 표현된다.
```

```python
s1.union(s2)       #union 함수를 사용
```

### 차집합 구하기

```python
s1 - s2   # - (빼기)를 사용하면 차집합을 구할 수 있다.
```

```python
s2 - s1
```

```python
# difference 함수를 사용
```

```python
s1.difference(s2)
```

```python
s2.difference(s1)
```

## 4.집합 자료형 관련 함수

### add() 값 1개 추가하기

```python
#이미 만들어진 set 자료형에 값을 추가할 수 있다. 1개의 값만 추가add할 때는 다음과 같이 하면 된다.
```

```python
s1 = set([1, 2, 3])
s1.add(4)
s1
```

### update() 값 여러개 추가하기

```python
#여러 개의 값을 한꺼번에 추가(update)할 때는 다음과 같이 하면 된다.
```

```python
s1 = set([1, 2, 3])
s1.update([4,5,6])
s1
```

### remove() 특정 값 제거하기

```python
# 특정 값을 제거하고 싶을 때는 다음과 같이 하면 된다
```

```python
s1 = set([1, 2, 3])
s1.remove(2)
s1
```
