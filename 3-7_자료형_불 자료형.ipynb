# 불 자료형 

```python
#불(bool) 자료형이란 참(True)과 거짓(False)을 나타내는 자료형이다. 불 자료형은 다음 2가지 값만을 가질 수 있다.
```

* True :참을 의미한다.
* False :거짓을 의미한다.

```python
#True나 False는 파이썬의 예약어로, true, false와 같이 작성하면 안 되고 첫 문자를 항상 대문자로 작성해야 한다.
```

## 1. 불 자료형은 어떻게 사용할까?

### 변수에 불 자료형 지정 

```python
a = True
b = False
```

### type 함수로 자료형 확인 

```python
# type(x)는 x의 자료형을 확인하는 파이썬의 내장 함수이다.
```

```python
type(a) #<class 'bool'>
```

```python
type(b) # #<class 'bool'>
```

### 불 자료형은 조건문의 리턴값으로도 사용 

```python
# 1==1 은 '1과1이 같은가?'를 묻는  조건문이다. 이런 조건문은 결과로 True 또는 False에 해당하는 불 자료형을 리턴한다.
```

```python
1==1
```

```python
#2는 1보다 크므로 2>1 조건문은 참이다 즉, True를 리턴한다.
#2는 1보다 작지 않으므로 2<1 조건문은 거짓이다. 즉, False를 리턴한다.
```

```python
2>1
```

```python
2<1
```

## 2. 자료형의 참과 거짓

```python
# 문자열, 리스트, 튜플, 딕셔너리 등의 값이 비어있으면 ("",[],(),{})거짓이 되고 비어 있지 않으면 참이 된다. 숫자에서는 그 값이 0일 떄 거짓이 된다. 
# 위 표를 보면 None은 거짓을 뜻한다는 것만 알아 두자.
```

### 자료형의 참과 거짓이 프로그램에서 어떻게 쓰이는지 간단히 알아보자.

```python
a = [1,2,3,4]
while a: print(a.pop())
```

```python
# while 조건문: 수행할_문장  /조건문이 참인 동안 조건문 안에 있는 문장을 반복해서 수행한다.
```

```python
if[]: print("참")    #[]는 앞의 표에서 볼 수 있듯이 비어 있는 리스트이므로 거짓이다. 따라서 "거짓"이라는 문자열이 출력된다.
else: print("거짓")
```

```python
if[1,2,3]: print("참")  #[1, 2, 3]은 요솟값이 있는 리스트이므로 참이다. 따라서 "참"을 출력한다
else: print("거짓")
```
