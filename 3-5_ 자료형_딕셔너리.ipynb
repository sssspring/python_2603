# 딕셔너리

## 1. 딕셔너리란?

```python
# 딕셔너리는 단어 그대로 '사전'이라는 뜻 . 딕셔너리는 key와 value를 한 쌍으로 가지는 자료형이다. 
```

## 2. 딕셔너리는 어떻게 만들까?

### 딕셔너리의 기본형

```python
#{Key1: Value1,Key2: Value2,Key3:Value3,...}
```

### 딕셔너리의 예

```python
dic = {'name': '이봄', 'phone': '010-9999-1234', 'birth': '0405'}
dic
```

```python
a = {1: 'hi'}  # Key로 정숫값 1, Value로 문자열 'hi'를 사용한 예이다.
```

```python
a
```

```python
a[1]
```

## 딕셔너리 쌍 추가, 삭제하기

### 딕셔너리 쌍 추가

```python
a = {1:'a'}       #a[2] = 'b'와 같이 입력하면 딕셔너리 a에 Key와 Value가 각각 2와 'b'인 {2: 'b'} 딕셔너리 쌍이 추가된다.
a[2] = 'b'
a
```

```python
a['name'] ='pey'
a
```

```python
a[3] = [1,2,3]
a
```

### 딕셔너리 요소 삭제하기

```python
del a[1]           #del 함수를 사용해서 del a[key]를 입력하면 지정한 Key에 해당하는 {Key: Value} 쌍이 삭제된다.
a
```

## 4. 딕셔너리를 사용하는 방법 

```python
# ‘딕셔너리는 주로 어떤 것을 표현하는 데 사용할까?’라는 의문이 들 것이다. 
#예를 들어 4명의 사람이 있다고 가정하고 각자의 특기를 표현할 수 있는 좋은 방법에 대해서 생각해 보자. 리스트나 문자열로는 표현하기가 매우 까다로울 것이다. 하지만 파이썬의 딕셔너리를 사용하면 이 상황을 표현하기가 정말 쉽다.
```

### 딕셔너리에서 Key를 사용해 Value 얻기

```python
grade = {'pey': 10, 'julliet': 99}
grade['pey']
```

```python
#리스트나 튜플, 문자열은 요솟값을 얻고자 할 때 인덱싱이나 슬라이싱 기법 중 하나를 사용했다. 하지만 딕셔너리는 단 1가지 방법뿐이다.
#'pey'라는 Key의 Value를 얻기 위해 grade['pey']를 사용한 것처럼 어떤 Key의 Value를 얻기 위해서는 '딕셔너리_변수_이름[Key]'를 사용해야 한다.
```

딕셔너리 만들 때 주의 사항

```python
a = {1:'a', 1:'b'}
a
```

```python
a = {[1,2]:'hi'}
a
```

## 5. 딕셔너리 관련 함수

### keys() key 리스트 만들기 

```python
a = {'name': '이봄', 'phone': '010-9999-1234', 'birth': '0405'}
a.keys()                                                     #a.keys()는 딕셔너리 a의 Key만을 모아 dict_keys 객체를 리턴
```

```python
list(dic.keys())
```

```python
for key in dic.keys():
    print(key)
```

### values() Value 리스트 만들기

```python
dic.values()
```

```python
for value in dic.values():
    print(value)
```

### items()  Key, Value 쌍 얻기

```python
dic.items()
```

### clear()   Key: Value 쌍 모두 지우기

```python
dic.clear()
```

```python
dic
```

### get()  Key로 Value 얻기

```python
a ={'name': '이봄', 'phone': '010-9999-1234', 'birth': '0405'}
```

```python
a['name']
```

```python
a.get('name')
```

```python
a['age'] = 30
a
```

```python
a.get('get',30)
```

### in 해당 Key가 딕셔너리 안에 있는지 조사하기

```python
#'name' 문자열은 a 딕셔너리의 Key 중 하나이다. 따라서 'name' in a를 호출하면 참(True)을 리턴한다. 이와 반대로 'email'은 a 딕셔너리 안에 존재하지 않는 Key이므로 거짓(False)을 리턴한다.
```

```python
a                             
```

```python
'name' in a
```

```python
'email' in a
```
