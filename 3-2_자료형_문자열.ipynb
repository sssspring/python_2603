## 7.문자열 포매팅

```python
# 1) 숫자 바로 대입
food = "i eat %d apples"%3
food
```

```python
# 2) 문자열 바로 대입
food = "i eat %s apples"%'three'
food
```

```python
# 3) 숫자 값을 나타내는 변수로 대입
number = 3
food = "i eat %d apples"% number
food
```

```python
# 4) 2개 이상의 값 넣기
num = 10
day = 'three'

'I eat %d apples. so I was sick for %s days'%(num, day)
```

```python
# %s는 숫자도 문자열로 대입 가능
'I have %s apples.' %3
```

```python
'I have %d apples.' %3.14 # %d 는 정수만 나타남
```

```python
'I have %f apples.'%3.14  # %f 는 소수점으로 보여지기 떄문에 소수점 아래 6자리가 다 나타남
```

```python
'I have %s apples.'%3.14  #숫자가 문자처럼 대입됨
```

```python
# 포맷 코드와 문자%를 같이 사용하려면 %%  (퍼센트 기호를 표현하고 싶을때)
'Error is %d%%.' %98
```

## 10.format() 함수를 사용한 포매팅

```python
# 숫자 바로 대입
'I eat {} apples.'.format(3)
```

```python
# 문자 바로 대입
'I eat {} apples.'.format('three')
```

```python
# 변수로 대입
num =3
'I eat {} apples.'.format(num)
```

```python
'I eat {0} apples.'.format('three') # 0 넣은것은 안넣은 값과 같다 
```

```python
# 2개 이상 값 넣기
num = 10 
day = 'three'
```

```python
'I eat {} apples. so I was sick for {} days.'.format(num, day)
```

```python
'I eat {0} apples. so I was sick for {1} days.'.format(num, day)
```

```python
'I eat {1} apples. so I was sick for {0} days.'.format(num, day)
```

```python
# 이름으로 넣기
'I eat {number} apples. so I was sick for {day} days.'.format(day=3, number=10)
```

```python
'I eat {} apples. so I was sick for {day} days.'.format(10, day=3) #인덱스와 이름을 혼용으로 넣기
```

```python
# 왼쪽 정렬 
'{:<10}'.format('hi')
```

```python
# 오른쪽 정렬 
'{:>10}'.format('hi')
```

```python
# 가운데 정렬 
'{:^10}'.format('hi')
```

```python
# 공백 채우기 
'{:=^10}'.format('hi')
```

```python
'{:!<10}'.format('hi')
```

```python
'{:!>10}'.format('hi')
```

```python
# 소수점 표현
'{:0.2f}'.format(3.428524)
```

```python
'{:10.2f}'.format(3.428524)
```

```python
'{:=^10.2f}'.format(3.428524)
```

```python
#{또는}문자표현
'{{and}}'.format()
```

```python
'{}'.format('{and}')
```

## 11. f 문자열 포매팅

```python
# 문자열 앞에 f접두사를 붙이기
name = '이봄'
age = 37
f'나의 이름은 {name}입니다. 나이는 {age}입니다'
```

```python
# 표현식 사용하기
f'나는 내년이면 {age+1}세가 됩니다'
```

```python
#딕셔너리
d = {'name': '이봄1', 'age' : 370}
f'나의 이름은 {d['name']}입니다. 나이는 {d['age']}입니다'
```

```python
# 정렬
f'{'hi':<10}'
```

```python
f'{'hi':<10}'
```

```python
f'{'hi':^10}'
```

```python
f'{'hi':=<10}'
```

```python
f'{'hi':=^10}'
```

```python
f'{'hi':!<10}'
```

```python
# 소수점 표현
f'{3.4284231:0.2f}'
```

```python
f'{3.4284231:10.2f}'
```

```python
f'{3.4284231:<10.2f}'
```
