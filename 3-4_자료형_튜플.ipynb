## 2. 튜플의 요솟값을 지우거나 변경하려고 하면 어떻게 될까?

### 튜플 요솟값을 삭제하려 할 때

```python
t1 = (1,2,'a','b')
```

```python
del t1[0]  #튜플은 요솟값을 지울 수 없다는 오류 메세지를 확인 할 수 있다.
```

### 튜플 요솟값을 변경하려 할 때

```python
t1 = (1,2,'a','b')
```

```python
t1[0]='c'    # 튜플의 요솟값을 면경하려고 해도 오류가 발생하는 것을 확인 할 수 있다.
```
